# Handle missing values

Fill Missing Values with median

```python
# Fill missing values in 'bedrooms' with the median
median_bedrooms = df['bedrooms'].median()
df['bedrooms'] = df['bedrooms'].fillna(median_bedrooms)

# Fill missing values in 'bathrooms' with the median
median_bathrooms = df['bathrooms'].median()
df['bathrooms'] = df['bathrooms'].fillna(median_bathrooms)

# Fill missing values in 'land_size_m2' with the median
median_land_size_m2 = df['land_size_m2'].median()
df['land_size_m2'] = df['land_size_m2'].fillna(median_land_size_m2)

# Fill missing values in 'building_size_m2' with the median
median_building_size_m2 = df['building_size_m2'].median()
df['building_size_m2'] = df['building_size_m2'].fillna(median_building_size_m2)

# Fill missing values in 'floors' with the median
median_floors = df['floors'].median()
df['floors'] = df['floors'].fillna(median_floors)
```

>> This code is used to handle missing values in numerical columns by filling them with the median of each column. The choice of median instead of mean is important because, as shown in the histograms earlier, many of these variables (bedrooms, bathrooms, land_size_m2, building_size_m2) are right-skewed with extreme outliers. Using the mean would pull the imputation toward unusually high values, while the median is more robust and better represents the “typical” property. This approach helps maintain data quality by ensuring no missing values remain, while also reducing the impact of outliers.

```python
# Fill missing values in categorical columns with 'Unknown'
df['property_condition'] = df['property_condition'].fillna('Unknown')
df['building_orientation'] = df['building_orientation'].fillna('Unknown')
df['furnishing'] = df['furnishing'].fillna('Unknown')
```

>> This code is used to handle missing values in categorical columns by replacing them with the placeholder "Unknown". Unlike numeric data, categorical variables (such as property_condition, building_orientation, and furnishing) cannot be imputed with mean or median, since they represent labels rather than quantities. By filling missing values with "Unknown", we preserve all rows in the dataset without introducing bias toward the most frequent category (mode imputation).

# Feature Engineering

```python
# Overwrite facilities with facility level (0–3)
if 'facilities' in df.columns:
    def facility_level(s):
        if pd.isna(s) or s.strip() == "":
            return 0
        # Split by comma, remove duplicates, count
        count = len(set([f.strip().lower() for f in s.split(',') if f.strip() != ""]))
        if count == 0:
            return 0
        elif count <= 3:
            return 1
        elif count <= 6:
            return 2
        else:
            return 3

    df['facilities'] = df['facilities'].apply(facility_level)

# Preview result
print(df['facilities'].unique())
```

>> This code is used to simplify and standardize the facilities column by converting long text strings of facility descriptions into a binned numeric scale (0–3). Instead of storing every facility name, the function counts how many unique facilities each property has, then groups (bins) the counts into levels: 0 = no facilities, 1 = few (1–3), 2 = medium (4–6), and 3 = many (7 or more). This is a form of feature engineering through binning, where continuous or variable-length text data is transformed into meaningful categories. The benefit is that the feature becomes much easier to analyze and use in models, since we now have a clear categorical variable that indicates the overall amenity level of a property rather than handling dozens of inconsistent facility strings.

```python
# Count the occurrences of each facility level
facility_counts = df['facilities'].value_counts().sort_index()

# Create the bar plot for facilities
plt.figure(figsize=(8, 6))
sns.barplot(
    x=facility_counts.index,
    y=facility_counts.values,
    hue=facility_counts.index,
    palette='viridis',
    legend=False
)
plt.xlabel('Facility Level', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('Distribution of Facility Levels', fontsize=14)

# Add labels to the bars
for i, v in enumerate(facility_counts.values):
    plt.text(i, v + 10, str(v), ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```

>> **Insight**: This bar plot shows the distribution of facility levels across properties. Most properties fall into `level 1 (few facilities)` with 1,349 listings, followed closely by `level 3 (many facilities)` with 1,235 listings, while `level 2 (medium facilities)` has the lowest count at 969. The insight is that the market is polarized: properties are often listed with either very few facilities or a large number of them, while moderate facility offerings are less common. This suggests that developers and sellers tend to differentiate their properties as either basic/affordable or premium/complete, with fewer options in the mid-range segment.

# Feature Engineering

```python
# Define bins and labels
bins = [0, 5, 10, 15, 20, 25, df['building_age'].max()]
labels = ['0-5 years', '6-10 years', '11-15 years', '16-20 years', '21-25 years', '>25 years']

# First, bin only numeric values
df['building_age'] = pd.cut(
    df['building_age'],
    bins=bins,
    labels=labels,
    include_lowest=True
)

# Add "Unknown" category and fill NaN
df['building_age'] = df['building_age'].cat.add_categories('Unknown')
df['building_age'] = df['building_age'].fillna('Unknown')

# Preview
print(df['building_age'].value_counts())
```

>> This code applies feature engineering via binning on `building_age` by grouping continuous ages into labeled ranges—0–5, 6–10, 11–15, 16–20, 21–25, and >25 years using `pd.cut()`. Turning raw numbers into these categories makes the data easier to interpret, reduces the impact of skew/outliers, and enables clearer comparisons (e.g., whether newer homes *0–5 years* command higher prices than older ones >25 years). In short, binning converts raw ages into meaningful segments that are more useful for visualization and modeling.
After binning, the code adds an explicit `Unknown` category and fills missing values with it. This preserves all rows (no data loss), keeps the column as a tidy categorical type, and clearly flags listings with unspecified ages rather than forcing a numerical imputation. The building_age column has a high proportion of missing values (around 40%, equal to 1,445 records). Instead of imputing these values with the median, which would artificially cluster many properties into one age category and bias the distribution, the missing values are replaced with "Unknown". This approach is more appropriate because building_age will later be binned into categorical ranges, and keeping "Unknown" as a separate category preserves data transparency without introducing false assumptions. It also allows for clearer analysis, as the impact of properties with unspecified ages can be directly observed rather than hidden within other groups.Finally, the original `building_age` column is intentionally overwritten so downstream analyses use the clean, categorical version.


```python
# Normalize to lowercase strings while preserving NaN
df['property_condition'] = df['property_condition'].astype('string').str.lower().str.strip()
df['furnishing'] = df['furnishing'].astype('string').str.lower().str.strip()

# Define furniture set
furniture_set = {'semi furnished', 'unfurnished', 'furnished'}

# Handle furnishing NaN
df.loc[df['furnishing'].isna() & df['property_condition'].isin(furniture_set),
       'furnishing'] = df['property_condition']

df.loc[df['furnishing'].isna() & ~df['property_condition'].isin(furniture_set),
       'furnishing'] = 'unknown'

# Replace 'baru' in furnishing with 'unknown'
df['furnishing'] = df['furnishing'].replace({'baru': 'unknown'})

# Replace misplaced furnishing values in property_condition
df['property_condition'] = df['property_condition'].replace({
    'unfurnished': 'unknown',
    'semi furnished': 'unknown',
    'furnished': 'unknown'
})

# Handle property_condition NaN
df.loc[df['property_condition'].isna() & (df['furnishing'] == 'baru'),
       'property_condition'] = 'baru'

df.loc[df['property_condition'].isna() & df['furnishing'].isin(furniture_set),
       'property_condition'] = 'unknown'

print(df['property_condition'].value_counts())
print(df['furnishing'].value_counts())
```

>> This code is designed to clean and standardize the property_condition and furnishing columns by resolving inconsistencies and handling missing values. Since some values such as unfurnished, semi furnished, and furnished appeared incorrectly in property_condition, they are reassigned to unknown to avoid overlap with the furnishing column. Missing values in furnishing are filled using related information from property_condition when possible, otherwise replaced with unknown. Additionally, baru entries in furnishing are corrected to unknown because they describe condition rather than furnishing. This cleaning process ensures that both columns represent their intended meaning clearly, property_condition focuses strictly on the state of the building (baru, bagus, renovasi, dll.), while furnishing consistently describes interior equipment (unfurnished, semi furnished, furnished).

```python
# Map values to scores
pc_map = {
    'baru': 3, 'bagus': 1, 'bagus sekali': 2,
    'sudah renovasi': 2, 'butuh renovasi': 0, 'unknown': 0
}
fz_map = {
    'unfurnished': 0, 'semi furnished': 1, 'furnished': 2, 'unknown': 0
}

# Create the combined score in a new column
df['condition_furnishing_score'] = (
    df['property_condition'].map(pc_map).fillna(0).astype(int) +
    df['furnishing'].map(fz_map).fillna(0).astype(int)
)

# Quick check
print(df[['property_condition', 'furnishing', 'condition_furnishing_score']].head())

print(df['condition_furnishing_score'].value_counts())
```

>> This code performs feature engineering using mapping to convert categorical values from `property_condition` and `furnishing` into numeric scores, then combines them into a single column for easier analysis. Two dictionaries (pc_map and fz_map) are defined to map each category to a score that reflects its relative quality—for example, baru is mapped to 3 (highest condition), bagus sekali and sudah renovasi to 2, while butuh renovasi and unknown get 0. Similarly, in furnishing, furnished is mapped to 2, semi furnished to 1, and both unfurnished and unknown to 0. These mapped values are then summed into a new feature called `condition_furnishing_score`, which represents the overall housing quality score by combining both the building’s physical condition and its furnishing status. This mapping transforms subjective categorical data into a structured numeric variable, making it easier to use in statistical analysis and machine learning models. It also provides a consolidated measure of property quality, where higher scores indicate better condition and more complete furnishing, which can be directly compared against property prices to analyze whether higher-quality homes yield higher market values.
